```python
# function to handle inconsistente values in round column
def handle_s_b (round):
    if round == 'SeriesB()':
        return "SeriesB"
    elif round == 'Private\\Equity':
        return 'PrivateEquity'
    elif round in ['Debt-Funding','DebtandPreferencecapital','StructuredDebt']:
        return 'DebtFunding'

    elif round in ['EquityBasedFunding','PrivateFunding','Private','Equity']:
        return 'PrivateEquity'
    elif round in ['SeriesC','SeriesD','SeriesF','SeriesE','SeriesG','SeriesH','SeriesJ']:
        return 'SeriesC+'
    elif round in ['Angel','AngelRound','Seedfunding','AngelFunding','SeedRound','SeedFunding','FundingRound','Seed/AngelFunding','Angel/SeedFunding','Seed\\Funding','SeedFundingRound','Seed/AngleFunding','Seed']:
        return "Seed/Angel"
    elif round in ["pre-seriesA",'Pre-seriesA','pre-SeriesA','PreSeriesA']:
        return 'Pre-SeriesA'
    else:
        return round
     

```

### Standardizing City Names Using Mapping

- Multiple city combinations (e.g., "bangalore / sfo", "new delhi / us")
- Different spellings and naming variations (e.g., "bengaluru", "gurgaon", "ahemdabad")
- International or irrelevant locations mixed with Indian cities (e.g., "india/us", "san francisco")
- Localities instead of cities (e.g., "andheri", "haryana")

- Created a custom mapping dictionary (`city_map`) to standardize values
